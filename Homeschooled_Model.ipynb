```python
m = larch.Model(d)
m.title = "Homeschooling"
#alternatives are:
#    1: Yes
#    2: No
 
```

```python
m.utility.co[1] = (
    + P('full time worker') * X('PERSTYPE==1')
    + P('part time worker') * X('PERSTYPE==2')
    + P('non-worker age [SSRetAge]') * X('PERSTYPE==3')
    + P('other non-worker') * X('PERSTYPE==4')
    + P('university student') * X('PERSTYPE==5')
    + P('driving age student') * X('PERSTYPE==6')
    + P('pre-driving- age student') * X('PERSTYPE==7')
    
    + P('CHILD_n_hh_children_sq_and_at_least_1_NWA') * X('(HHNWA>0) * ((HHCU5+HHCU515)*(HHCU5+HHCU515)) * (PERSTYPE in (7,8))')
    + P('ADULT_n_hh_children_sq_and_at_least_1_NWA') * X('(HHNWA>0) * ((HHCU5+HHCU515)*(HHCU5+HHCU515)) * (PERSTYPE not in (7,8))')

    + P('low_modest_income_hh') * X('INCOME<=6')
    + P('medium_income_hh') * X('INCOME in (7,8)')
    + P('upper_income_hh') * X('INCOME IN (9,10)')
    + P('top_income_hh') * X('INCOME==11')
    + P('miss_income_hh') * X('INCOME==99')

    + P('PPACOG') * X("PPACOG")
 )

m.parameter('full time worker', value=-99, holdfast=True)
m.parameter('part time worker', value=-99, holdfast=True) 
m.parameter('non-worker age [SSRetAge]', value=-99, holdfast=True) 


m.maximize_loglike()

print(m)
# LL to compare -20254.811821313644
```

```python
m.blurb = """
This model represents a binary choice model for home-schooling.  For every person
who is a full or part time student, the choice is to be home-schooled (includes all school
activities at home including online coursework) or not.  If not, the person will then be
dispatched to the usual school location model for the choice of an out-of-home school location.

All students are included in this model, regardless of age.

Input data includes:

    * person type
    
    * household makeup

      - number of children under 16
      - number of non-working adults

    * household income (standard categories)
    
    * whether the home location is in PPACOG (which has
      a higher rate of homeschooling than elsewhere in the state).

"""

m.xhtml(
    '!',
    'data',
    'datasummary', 
    filename=os.path.expanduser('~/CamSys/ColoradoDOT/results/report-{}.html'.format(m.title.casefold().replace(' ','-')))
)
```
